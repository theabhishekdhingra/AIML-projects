## About the dataset

This is perhaps the best known database to be found in the pattern recognition literature. The data set contains 3 classes of 50 instances each, where each class refers to a type of iris plant.  One class is linearly separable from the other 2; the latter are NOT linearly separable from each other.

Attribute Information:
   1. sepal length in cm
   2. sepal width in cm
   3. petal length in cm
   4. petal width in cm
   5. class: 
      -- Iris Setosa
      -- Iris Versicolour
      -- Iris Virginica

### Write a function that accepts two numbers as input and prints them - Pass the Sepal length and sepal width of 5th row and print the output

```python
def input_print(sl, sw):             # Recieving parameters "sepal lenght" and "sepal width"
    print("Sepal Length is {}".format(sl))    #printing sepal length
    print("Sepal width is {}".format(sw))     #printing sepal width

input_print(Iris.loc[5,"Sepal Length (in cm)"], Iris.loc[5,"Sepal Width in (cm)"] )   #passing arguments
```

### Find the range of all the columns in the dataset

*Range = Max value - Min value (in the column)*

```python
print("Range of Sepal lenght is {}".format(Iris["Sepal Length (in cm)"].max() - Iris["Sepal Length (in cm)"].min()) )
print("Range of Sepal width is {}".format(Iris["Sepal Width in (cm)"].max() - Iris["Sepal Width in (cm)"].min()) )
print("Range of Petal length is {}".format(Iris["Petal length (in cm)"].max() - Iris["Petal length (in cm)"].min()) )
print("Range of Petal width is {}".format(Iris["Petal width (in cm)"].max() - Iris["Petal width (in cm)"].min()) )
```
