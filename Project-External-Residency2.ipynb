```python
import pandas as pd
import numpy as np
```

```python
laliga=pd.read_csv('Laliga.csv',header=1)
```

```python
laliga.sample(10).T
```

```python
lalinga_col=laliga.columns
laliga.shape
```

```python
laliga.replace(to_replace='-',value=0,inplace=True)
laliga.head()
```

```python
laliga['startyear']=laliga.Debut.str.split('-',n=1,expand=True)[0]
laliga['startyear']=pd.to_numeric((laliga['startyear']),downcast='signed')
laliga['Since/LastApp']=np.where(laliga['Since/LastApp'] == '1929',"2017-18",laliga['Since/LastApp'])
laliga['LastPlayed']=laliga['Since/LastApp'].str.split('-',n=1,expand=True)[0] 
laliga['LastPlayed']=pd.to_numeric((laliga['LastPlayed']),downcast='signed')
laliga['LastPlayed']=laliga['LastPlayed'] + 1
laliga.head(10)
```

```python
print (laliga.loc[(laliga.startyear>1930) & (laliga.startyear<1980)])
```

```python
laliga['Points']=pd.to_numeric((laliga['Points']),downcast='signed')
laliga.sort_values(by=['Points'],ascending=False).head(5)
```

```python
def Goal_diff_count(laliga_pass):
    gf=pd.to_numeric((laliga_pass['GoalsFor']),downcast='signed')
    ga=pd.to_numeric((laliga_pass['GoalsAgainst']),downcast='signed')
    laliga['GoalDiff']=pd.Series(gf-ga)
    print("Maximum goal difference is {}".format(max(laliga_pass['GoalDiff'])))
    print("Minimum goal difference is {}".format(min(laliga_pass['GoalDiff'])))
   
Goal_diff_count(laliga)
```

```python
laliga['GamesPlayed']=pd.to_numeric((laliga['GamesPlayed']),downcast='signed')
laliga['GamesWon']=pd.to_numeric((laliga['GamesWon']),downcast='signed')
laliga['WinningPercent']=(laliga['GamesWon']/laliga['GamesPlayed'])*100
laliga['WinningPercent'].fillna(value='0',inplace=True)
laliga
```
