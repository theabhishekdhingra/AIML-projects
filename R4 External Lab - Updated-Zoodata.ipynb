### Q7. Find out the linkage matrix

### Q8. Plot the dendrogram.

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,6))
dendogram=sch.dendrogram(sch.linkage(zoo_df,method='average'))
plt.title("Dendogram on Zoo")
plt.xlabel("Animal_Number")
plt.ylabel("Animal_Class")
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,6))
dendogram=sch.dendrogram(sch.linkage(zoo_df,method='single'))  # Distance as single
plt.title("Dendogram on Zoo")
plt.xlabel("Animal_Number")
plt.ylabel("Animal_Class")
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,6))
dendogram=sch.dendrogram(sch.linkage(zoo_df,method='complete'))  # disatance as complete
plt.title("Dendogram on Zoo")
plt.xlabel("Animal_Number")
plt.ylabel("Animal_Class")
```

### Q9. Recreate the dendrogram for the last 4, 7, and 12 merged clusters 

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,6))
dendogram=sch.dendrogram(sch.linkage(zoo_df,method='average'),p=4,truncate_mode='lastp')
plt.title("Dendogram on Zoo")
plt.xlabel("Animal_Number")
plt.ylabel("Animal_Class")
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,6))
dendogram=sch.dendrogram(sch.linkage(zoo_df,method='average'),p=7,truncate_mode='lastp')
plt.title("Dendogram on Zoo")
plt.xlabel("Animal_Number")
plt.ylabel("Animal_Class")
```

```python
import scipy.cluster.hierarchy as sch
plt.figure(figsize=(15,6))
dendogram=sch.dendrogram(sch.linkage(zoo_df,method='average'),p=12,truncate_mode='lastp')
plt.title("Dendogram on Zoo")
plt.xlabel("Animal_Number")
plt.ylabel("Animal_Class")
```
